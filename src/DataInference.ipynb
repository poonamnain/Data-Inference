# Q2. What is an appropriate set of hypothesis for this task? What kind of statistical test do you expect to perform? Justify your choices.

We can do Null and Alternative hypothesis.

Null Hypothesis- There would be a lower mean time for the incongruent tests than the congruent tests or no
noticeable difference in time duration between the congruent and incongruent tests
This is the result that we want to provide evidence against. This is always an equality and is denoted by  H0. In our case for null hypothesis we would assume that, the change in the color of the words doesn't impact the time taken in identifying them. We would prove it wrong.

H0:μ=0

Alternative Hypothesis- There would be a noticeable increase in time duration between the congruent and incongruent
tests. 
This is a result that we are testing against our null hypothesis claim. This value is always an inequality and is denoted by  Ha. 

Ha:μ≠0


This experiment should be a single-tailed type scenario as we see that the response time always increases in case of incongruent, when compared to congruent response times. It never decreases.

• µc = the mean of time spent for the participant group of congruent test
• µi = the mean of time spent for the participant group of incongruent test

The null hypothesis (H0):
µc - µi ≥ 0, the time duration difference of the congruent and incongruent tests would be equal to 0 or be greater than 0, meaning the incongruent times would be less or equal to the congruent times for the population.

The alternative hypothesis (HA):
µc - µi < 0, the time duration difference of the congruent and incongruent tests would be less than 0, meaning the incongruent times would be greater than the congruent test times for the population.

The recommended statistical test we should use for the Stroop experiment is a T-test as we don’t know the mean of the total population however we have detailed data on a sampled group for both the congruent and incongruent tests. In addition, we only have a data set of 24 people in the sample, making T-test a better choice to do our statistical test.

The type of T-Test selected is:
• Dependent: I chose to go the dependent t-test route as even though we do two tests
(congruent and incongruent) it is with the same group of people that was tested on the
congruent words then tested on the incongruent words.
• Paired: I chose to go the paired t-test because the participants were measured at two
time points (congruent and incongruent word sets).

# Q5. Now, perform the statistical test and report your results. What is your confidence level and your critical statistic value? Do you reject the null hypothesis or fail to reject it? Come to a conclusion in terms of the experiment task. Did the results match up with your expectations?


As determined in question 2 we will calculate our data using the T-Test statistical test for the Stroop
Effect experiment. Considering the group took both the congruent and incongruent tests this T-Test
will be considered dependent and paired.


Knowing the means of each test congruent and incongruent from above
mean of congruent µc = 14.05
mean of incongruent µi = 22.02

We can calculate quickly a point estimate using: µc - µi
Point_estimate = µc - µi = 14.05-22.02 
= -7.97

A point estimate is handy to know as it gives us our “best guess” on the difference of the population
between the two tests.

Mean of differences dbar = sigma of difference of congruent values and incongruent values divided by n
= 7.96

We can see that point_estimate is almost equal to mean of differences.

Let's calculate the stantard deviation of the differences value
--------------------------------------------------------------

Using the average of the differences which should match the point estimate of -7.97 we subtract that value
from the differences of each person and then square it.

Having this set of squared deviations from the mean we can sum them up into Square of Sums Using
∑sqr(𝒙𝒊 − 𝒙bar) for i = 1 to 24
= 544.33

Given the Square of Sums, lets call it SS, we can use this to determine the variance of the difference. Since we are
dealing with a sample we need to use the equation with the denominator of n-1, n being the number of
participants on the given tests (n = 24)

variance of difference = (SS)/(n-1)
= 544.33/23
= 23.67

Given the Variance for the difference we can use that knowledge to get the standard deviation of differences.

Standard deviation of differences = sqrt(variance of differences)
=  sqrt(23.67)
= 4.87

Given SD of differences, lets call it 's', we can now calculate the T-statistics for the differences.

T-statistics
-----------------------------------------------------------------------

T_stat = (mean of congruent - mean of incongruent)/ (s).(sqrt(n))

Given the numerator is the point estimate or average of the differences, and “n” is the number of participants in the Stroop experiment we can calculate the t_stat as:

=  -7.97 / (4.87)sqrt(24)
= -8.026
T_stat = -8.026

Since we have the T_stat we can compare it to the T_critical value to determine if the null hypothesis
should be accepted or rejected. 

We know there are 24 participants, which means degree of freedom df = (n-1) = 23

Since we are aiming to find with 99% Confidence level for one tail our α = 0.01. The critical statistic value for a one-tailed test in the negative direction is -2.5 

T_critical = -2.5 (from the critical T-Table)

Since the T_stat is in the critical region, the null hypothesis is rejected.

The t-test confirms that incongruent tests would normally take longer than congruent tests. 
The congruent test linked both visual colors with the correct printed name of the color, while incongruent required to try to separate the visual color and the name of the color which required more response time.

Conclusion:

I reject the null hypothesis, the color of the printed word does cause a greater time response.
The results match my expectations.
